```python
from hakai_api import Client
import hakai_profile_qc.review

# Setup Hakai-api Client
client = Client()
```

```python
# Show on one hakai_id
hakai_id = ["65679_2021-05-19T03:06:48.833Z"]
api_root = "https://goose.hakai.org/api"
```

```python
# Run Test on Hakai_ID
df = hakai_profile_qc.review.run_tests(hakai_id=hakai_id,api_root=api_root,drop_single_test=False)
df.head()
```

```python
#  Index by ctd_data_pk, consider only flag columns to be updated and convert to a dictionary
dict_output = (
    df.set_index("ctd_data_pk")
    .filter(regex="_flag$|_flag_level_1")
    .drop("direction_flag", axis="columns")
    .to_dict(orient="index")
)
dict_output
```

```python
# Use Hakai python api client
endpoint = "ctd/views/file/cast/data"
url = f"{api_root}\{endpoint}"

# Upload data through the API to the database
response = client.put(url, dict_output)
```
